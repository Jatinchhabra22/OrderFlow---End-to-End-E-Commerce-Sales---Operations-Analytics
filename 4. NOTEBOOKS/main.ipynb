```python
## 2) Managing the status of delivery using eda 

def map_status(x):
    x = str(x).lower()
    
    if "shipped - delivered" in x or "shipped - picked" in x or "out for delivery" in x or x == "shipped":
        return "Shipped"

    elif "returned" in x or "rejected" in x or "cancelled" in x or "damaged" in x:
        return "Returned/Cancelled"

    elif "pending" in x or "waiting" in x or "shipping" in x:
        return "Pending"
    
    else:
        return "Other"
```

```python
dataset['Status-Grouped'] = dataset['Status'].apply(map_status)
```

```python
dataset['Status-Grouped'].value_counts()
```

```python
dataset.head()
```

```python
dataset['Status'] = dataset['Status-Grouped']
dataset.drop(columns=['Status-Grouped'], inplace=True)
```

```python
## 3) Providng flag values as 0/1 for succesfull/unsuccessfull shipment 

dataset['Sucesss'] = dataset['Status'].apply(lambda x : 1 if "Shipped" in str(x) else 0)
```

```python
## 4)  now lets catogirze the sizes of dress here 

## s/m/l --> regular

## xl/xxl,3xl,4xl,5xl,6xl,freee --> oversized

def fit(x):
    x = str(x).upper()   # convert to uppercase for safety
    
    if "XL" in x:        # 2XL,3XL,4XL also captured
        return "Oversized"
    elif "L" in x:       # L captured
        return "Regular"
    elif "M" in x:
        return "Regular"
    elif "S" in x:
        return "Regular"
    else:
        return "Unknown"
        
dataset["Size_group"] = dataset["Size"].apply(fit)

```

```python
### 5) now time for some more features that help visualise better 

def price_bucket(x):
    if x < 300: return "Budget"
    elif x < 700: return "Mid Range"
    else: return "Premium"

dataset['Price_Segment'] = dataset['Amount'].apply(price_bucket)
```
